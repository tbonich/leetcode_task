```python
n = 2
m_lin = 1
m_nonlin = 1
con = 1 << n
name_in_lin = "func_lin_" + str(n) + ".txt"
name_in_nonlin = "func_nonlin_" + str(n) + ".txt"
ll_lin = []
ll_nonlin = []
name_out = "res_" + str(n) + ".txt"
name_f = "f_" + str(n) + ".txt"
name_state = "func_state_" + str(n) + ".txt"

with open(name_in_lin, "r") as file:
    for i in range(m_lin):
        ll_lin.append(file.readline().rstrip())
        
with open(name_in_nonlin, "r") as file:
    for i in range(m_nonlin):
        ll_nonlin.append(file.readline().rstrip())

list_of_gamma_lin = [[] for i in range(m_lin)]
list_of_gamma_nonlin = [[] for i in range(m_nonlin)]

for i in range(1, con):
    iv = i
    for j in range(m_lin):
        gamma = ""
        v = iv
        for t in range(con):
            gamma += str(v >> (n - 1))
            v = ((v << 1) ^ ((v >> (n - 1)) << n)) + int(ll_lin[j][v])
            if v == iv or (len(gamma) == con - 1):
                break
        list_of_gamma_lin[j].append(gamma)

for i in range(1, con):
    iv = i
    for j in range(m_nonlin):
        gamma = ""
        v = iv
        for t in range(con):
            gamma += str(v >> (n - 1))
            v = ((v << 1) ^ ((v >> (n - 1)) << n)) + int(ll_nonlin[j][v])
            if v == iv or (len(gamma) == con - 1):
                break
        list_of_gamma_nonlin[j].append(gamma)


min_i = con - 1
max_i = 0
cur_i = con - 1
max_i_f = 0
min_i_f = con - 1
st = ""
sup_min_i = con - 1
sup_max_i = 0

print("start ==========================================================")

with open(name_out, "w") as file:
    for i in range(m_nonlin):
        max_i_f = 0
        min_i_f = con - 1
        st = ""
        file.write(ll_nonlin[i] + ":" + '\n')
        for t in range(con - 1):
            s = ""
            min_i = con - 1
            max_i = 0
            cur_i = con - 1
            file.write(list_of_gamma_nonlin[i][t][:n] + '\n')
            for j in range(m_lin):
                if list_of_gamma_nonlin[i][t] != list_of_gamma_lin[j][t]:
                    cur_i = con - 1
                    s_nonlin = ""
                    s_lin = ""
                    for k in range(con - 1):
                        if s_nonlin == s_lin:
                            cur_i -= 1
                            s_nonlin += list_of_gamma_nonlin[i][t][k]
                            s_lin += list_of_gamma_lin[j][t][k]
                        else:
                            cur_i += 1
                            break
                    s += str(cur_i) + " (" + ll_lin[j] + ") "
                    file.write(s)
                    s = ""
                    min_i = min(min_i, cur_i)
                    max_i = max(max_i, cur_i)
                # else:
                    # file.write("0 (" + ll_lin[j] + ") ")
            s += "min_i = " + str(min_i) + " max_i = " + str(max_i)
            file.write(s + '\n')
            s = ""
            max_i_f = max(max_i_f, max_i)
            min_i_f = min(min_i_f, min_i)
        st = "min_i_f = " + str(min_i_f) + " max_i_f = " + str(max_i_f)
        file.write(st + '\n')
        sup_min_i = min(sup_min_i, min_i_f)
        sup_max_i = max(sup_max_i, min_i_f)
    file.write('sup_min = ' + str(sup_min_i) + ' sup_max = ' + str(sup_max_i) + '\n')


# with open(name_out, "w") as file:
#     file.write('sup_min = ' + str(sup_min_i) + ' sup_max = ' + str(sup_max_i) + '\n')
#     for i in range(m):
#         for el in list_of_i[i]:
#             file.write(el + '\n')

# with open(name_f, "w") as file:
#     for el in ll:
#         file.write(el + '\n')
        
# print("mid ==========================================================")

# with open(name_state, "w") as file:
#     for i in range(m):
#         file.write(ll[i] + ": " + '\n')
#         for el in list_of_gamma[i]:
#             file.write(el + ' ')
#         file.write('\n')
print("end ==========================================================")
```
